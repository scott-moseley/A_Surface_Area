```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import seaborn as sns

excel_file = 'cKWSA_Prepared.xlsx'
SA_Daily_DS2 = pd.read_excel(excel_file, sheet_name=1, index_col=0)
SA_Daily_DS2.head()
```

```python
X = SA_Daily_DS2[['Furnace Temp','HV Temp','D4 Temp']]
y = SA_Daily_DS2['Average SA']
```

```python
X_Matrix = X.values
X_Matrix
y_Matrix = y.values
```

```python
model = sm.OLS(y_Matrix,X_Matrix).fit
```

```python
model.summary()
```

```python
correlation_matrix = np.corrcoef(X_Matrix)
sns.heatmap(correlation_matrix)
```
